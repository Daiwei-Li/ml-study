```python
#Copy Train data excluding target
trainData_copy = trainData.iloc[:,:-1].copy()
testData_copy = testData.copy()
y_train= trainData.iloc[:,-1:]

#X_Test.head()
#X_Train.head()
```

```python
#List of Categorical Features
categoricalFeatures = [ 'MSSubClass', 'MSZoning', 'Street', 'Alley', 'LotShape', 'LandContour', 'Utilities', 'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'Condition2', 'BldgType', 'HouseStyle',  'RoofStyle', 'RoofMatl', 'Exterior1st', 'Exterior2nd', 'MasVnrType', 'ExterQual', 'ExterCond', 'Foundation', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1',  'BsmtFinType2', 'Heating', 'HeatingQC', 'CentralAir', 'Electrical',   'BsmtFullBath', 'BsmtHalfBath', 'FullBath', 'HalfBath', 'BedroomAbvGr', 'KitchenAbvGr', 'KitchenQual', 'TotRmsAbvGrd', 'Functional', 'Fireplaces', 'FireplaceQu', 'GarageType', 'GarageYrBlt', 'GarageFinish', 'GarageCars', 'GarageQual', 'GarageCond', 'PavedDrive', 'PoolQC', 'Fence', 'MiscFeature', 'MoSold', 'YrSold', 'SaleType', 'SaleCondition']


combined_Data = pd.concat([trainData_copy,testData_copy], keys=[0,1])
combined_Data = combined_Data.drop(['Id', 'LotFrontage', 'LotArea','OverallQual', 'OverallCond','YearBuilt', 'YearRemodAdd',  'MasVnrArea','BsmtFinSF1','BsmtFinSF2','BsmtUnfSF',   'TotalBsmtSF','1stFlrSF', '2ndFlrSF','LowQualFinSF', 'GrLivArea','GarageArea', 'WoodDeckSF', 'OpenPorchSF', 'EnclosedPorch', '3SsnPorch', 'ScreenPorch', 'PoolArea', 'MiscVal', ],axis=1)


#Do one Hot encoding for categorical features
combined_Data = pd.get_dummies(combined_Data,columns=categoricalFeatures)

#Separate Train data and test data
X_Train = combined_Data.xs(0)
X_Test = combined_Data.xs(1)

print(X_Train.head())
print(X_Test.head())
```

```python
    print(X_Train.shape)
print(X_Test.shape)
```

```python
#Decision Tree Classifier ========================================================================

clf = DecisionTreeRegressor()
clf.fit(X_Train, y_train)
clf_predict=clf.predict(X_Test)
X_Pred_train = clf.predict(X_Train)
y_pred_test = clf.predict(X_Test)
#Model Accuracy
print("TRAIN Accuracy:", accuracy_score(y_train,X_Pred_train))


pred=pd.DataFrame(clf.predict(X_Test),columns=["SalePrice"])
#pred['Id']=testData['Id']
#pred=pred[['Id','SalePrice']]
pred=pred[['SalePrice']]
pred.head()
```
